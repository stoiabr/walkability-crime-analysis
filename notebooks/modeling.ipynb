```python
import numpy as np
import pandas as pd

file_path = "/Users/brandonstoia/Desktop/MADS/Term 3/Session 1/SIADS 696 - Milestone II/ICPSR_38795/DS0001/38795-0001-Data.tsv"

df = pd.read_csv(
    file_path,
    sep="\t",
    low_memory=False,
    usecols=range(77)
)

cols = df.columns[33:77]
jcols = df.columns[22:31]

df[cols] = (
    df[cols]
        .astype(str)
        .apply(lambda x: x.str.strip().str.replace(" ", "", regex=False))
        .replace({"": np.nan, "99999": np.nan, "99998": np.nan})
        .apply(pd.to_numeric, errors="coerce")
)

df["TOTAL"] = df[cols].sum(axis=1, skipna=True)
df['OFFENSE'] = df['OFFENSE'].astype(str).str.zfill(3).replace({"018": "18", "019": "19"})
df = df.drop(columns=list(jcols) + list(cols))

df = (
    df
        .groupby(["ORI", "OFFENSE"], as_index=False)["TOTAL"]
        .sum()
)

df
```

```python
# normalize agc ori: strip trailing zeros
agc["ORI_norm"] = agc["ori"].str.strip().str.upper().str.rstrip("0")

# normalize crime ORI: uppercase + trim
crime["ORI_norm"] = crime["ORI"].astype(str).str.strip().str.upper()

# quick sanity checks
agc[["ori","ORI_norm"]].head(), crime[["ORI","ORI_norm"]].head()
```

```python
merged = agc.merge(crime, how="left", left_on="ORI_norm", right_on="ORI_norm")
```

```python
merged
```

```python
agc_only = set(agc["ORI_norm"]) - set(crime["ORI_norm"])
crime_only = set(crime["ORI_norm"]) - set(agc["ORI_norm"])

len(agc_only), len(crime_only)
```

```python
# keys
agc["ORI9"] = agc["ori"].astype(str).str.strip().str.upper()
crime["ORI7"] = crime["ORI"].astype(str).str.strip().str.upper()

# pad crime ORI to 9 chars (right-pad zeros)
crime["ORI9"] = crime["ORI7"].str.ljust(9, "0")

# unmatched counts using ORI9<->ORI9
agc_only = set(agc["ORI9"]) - set(crime["ORI9"])
crime_only = set(crime["ORI9"]) - set(agc["ORI9"])
len(agc_only), len(crime_only)
```

```python
merged = agc.merge(crime, how="inner", left_on="ORI9", right_on="ORI9")
merged
```
